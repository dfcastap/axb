# R

A quick look at the general method...

```r
# Using inverse...
A <- matrix(c(1,2,3,4), ncol=2)
A
```

```r
b <- c(7, 10)
b
```

```r
x <- solve(A)%*%b
x
```

```r
# Or...
solve(A, b)
```

```r
A %*% x - b
```

A non-square matrix...

```r
m <- 10; n <- 5
A <- matrix (runif (m*n), m, n)
```

```r
A
```

```r
b <- matrix (runif (m), m, 1)
b
```

```r
x <- qr.solve(A, b)
x
```

```r
A %*% x
```
